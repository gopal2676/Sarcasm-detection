```python
dataset = []
with open('Sarcasm_Headlines_Dataset.json','r') as js:
    for line in js:
        dataset.append(json.loads(line))
```

```python
article_link = []
headline = []
is_sarcastic = []
for data in dataset:
    article_link.append(data['article_link'])
    headline.append(data['headline'])
    is_sarcastic.append(data['is_sarcastic'])
```
